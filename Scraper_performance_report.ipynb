```python
# changes made previous code excluded parents in the df after line line 25
def children_matcher(matched_df, full_df, parent_slug, parent_df):
    matched_children = matched_df[((matched_df['slug'].str.startswith(f"{parent_slug}")) & (matched_df['new_slug'] != parent_slug))]
    unmatched_children = full_df[((full_df['slug'].str.startswith(f"{parent_slug}")) & (full_df['new_slug'] != parent_slug))]
    difference = unmatched_children.shape[0] - matched_children.shape[0]
    parent_df.loc[len(parent_df)] = [matched_children.shape[0], difference, parent_slug]
    return parent_df
```

```python
location_df['new_slug'] = location_df['slug'].str.split('-').str[0]
location_df.head(2)
```

```python
locations_that_matched = location_df[location_df['name'].isin(df['property_district'])] #re-inintialized to include all location df newly formed columns
parent_df = pd.DataFrame(columns=['matched', 'unmatched', 'slug'])
location_parents = list(location_df['new_slug'])
for slug in location_parents:
    parent_df = children_matcher(locations_that_matched, location_df, slug, parent_df)
```

```python
parent_df['id'] = range(1, len(parent_df) +1)
parent_df['combined'] = parent_df['unmatched'] + parent_df['matched']
parent_df['adCount'] = parent_df['slug'].map(location_df.set_index('new_slug')['adCount'])
parent_df['actual_listings'] = parent_df['slug'].map(property_search.set_index('slug')['externalID'].copy())
```
